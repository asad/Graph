# Comprehensive Graph Analysis: Use Cases, Algorithms, and Applications

Load points are defined as hot spots in the metabolic network (enzymes/metabolites) based on the ratio of number of k-shortest paths passing through a metabolite/enzyme (in/out) and number of nearest neighbour links (in/out) attached to it, compared to the average load value in the network.


# Graph Algorithms for Choke Points, Load Points, and Centrality Analysis
---
## **1. Choke Points**
### **1.1 Edge-Based Choke Points**
- **Definition**: Edges whose removal disrupts connectivity or significantly increases path lengths in the graph.
- **Mathematical Representation**:
 For an edge $e = (u, v)$, the cost impact is:
 $$
 \text{Cost}(e) = \frac{\text{New Path Length} + \epsilon}{\text{Original Path Length} + \epsilon}
 $$
 Where:
 - **New Path Length**: The shortest path between $u$ and $v$ after removing $e$.
 - $\epsilon$: A small constant to avoid division by zero.
---
### **1.2 Node-Based Choke Points**
- **Definition**: Nodes connected by exactly two edges. Removing these edges isolates the node or forces longer alternative paths.
- **Detection**:
 A node $n$ is a choke point if:
 - $\text{Degree of } n = 2$, and
 - Removal of $n$ disrupts or elongates paths between its neighbors.
---
## **2. Load Points**
### **2.1 Load Algorithm**
- **Definition**: Identifies nodes acting as traffic hotspots in the network, based on **k-shortest paths** and nearest neighbors.
- **Mathematical Representation**:
 $$
 L_{m(\text{in/out})} = \ln \left[ \frac{\frac{p_{m(\text{in/out})}}{k_{m(\text{in/out})}} + \epsilon}{\frac{\sum_{i=1}^M P_{i(\text{in/out})}}{\sum_{i=1}^M K_{i(\text{in/out})}} + \epsilon} \right]
 $$
 Where:
 - $p_{m(\text{in/out})}$: Number of k-shortest paths passing through node $m$.
 - $k_{m(\text{in/out})}$: Number of nearest neighbors (inbound or outbound) for $m$.
 - $P_{i(\text{in/out})}$: Total k-shortest paths across all nodes.
 - $K_{i(\text{in/out})}$: Total nearest neighbors across all nodes.
 - $\epsilon$: A small constant to avoid division by zero.
---
## **3. Centrality Measures**
### **3.1 Degree Centrality**
- **Definition**: Measures the number of direct connections of a node relative to the total number of nodes.
- **Mathematical Representation**:
 $$
 C_D(v) = \frac{\text{deg}(v)}{N - 1}
 $$
 Where:
 - $\text{deg}(v)$: Degree of node $v$.
 - $N$: Total number of nodes.
---
### **3.2 Betweenness Centrality**
- **Definition**: Measures the importance of a node based on how many shortest paths pass through it.
- **Mathematical Representation**:
 $$
 C_B(v) = \sum_{s \neq v \neq t} \frac{\sigma_{st}(v)}{\sigma_{st}}
 $$
 Where:
 - $\sigma_{st}$: Total number of shortest paths between nodes $s$ and $t$.
 - $\sigma_{st}(v)$: Number of those paths that pass through $v$.
---
### **3.3 Closeness Centrality**
- **Definition**: Measures how close a node is to all other nodes in the network.
- **Mathematical Representation**:
 $$
 C_C(v) = \frac{1}{\sum_{u \neq v} d(u, v)}
 $$
 Where:
 - $d(u, v)$: Shortest path distance between nodes $u$ and $v$.
---

## **Pros and Cons**
| **Algorithm**           | **Pros**                                                                                     | **Cons**                                                                                     |
|--------------------------|---------------------------------------------------------------------------------------------|---------------------------------------------------------------------------------------------|
| **k-Shortest Paths**     | Finds multiple paths for redundancy and alternative routes.                                 | Computationally expensive for large \( k \) and complex graphs.                            |
| **Load Algorithm**       | Highlights traffic hotspots by integrating global and local features.                      | Requires accurate path and neighbor link data; computationally intensive.                  |
| **Degree Centrality**    | Simple, fast, and identifies hubs effectively.                                              | Ignores global context and may be misleading in sparse or heterogeneous networks.          |
| **Betweenness Centrality** | Detects critical nodes that act as bridges in the network.                                | High computational cost; less sensitive to local dynamics.                                 |
| **Closeness Centrality** | Highlights nodes close to others, minimizing latency.                                       | Sensitive to disconnected graphs and ignores edge weights.                                 |
| **Choke Points**         | Detects critical edges and single points of failure.                                        | Computationally intensive for large networks; focuses only on edges, not nodes.            |

```python
import heapq
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import pandas as pd
```

```python

EPSILON = 1e-12  # Small constant to avoid infinities
# 1. k-Shortest Paths
def k_shortest_paths(graph, source, target, k=3, weight='weight'):
   paths = []
   queue = [(0, source, [])]  # (cost, node, path)
   while queue and len(paths) < k:
       cost, node, path = heapq.heappop(queue)
       path = path + [node]
       if node == target:
           paths.append((cost, path))
           continue
       for neighbor, edge_data in graph[node].items():
           if neighbor not in path:
               heapq.heappush(queue, (cost + edge_data.get(weight, 1), neighbor, path))
   return paths
# 2. Path-Based Metrics for Load Algorithm
def compute_path_counts(graph, k=3):
   p_in, p_out = defaultdict(int), defaultdict(int)
   for source in graph.nodes:
       for target in graph.nodes:
           if source != target:
               paths = k_shortest_paths(graph, source, target, k=k)
               for _, path in paths:
                   for node in path:
                       if node != source:
                           p_in[node] += 1
                       if node != target:
                           p_out[node] += 1
   return dict(p_in), dict(p_out)
# 3. Load Algorithm
def compute_load(graph, p_in, k_in, p_out, k_out):
   """
   Computes the load for all nodes in the graph.
   Parameters:
       graph: A NetworkX graph.
       p_in, p_out: Path counts for each node.
       k_in, k_out: Neighbor counts for each node.
   Returns:
       Dictionary with inbound and outbound loads.
   """
   total_p_in, total_k_in = sum(p_in.values()), sum(k_in.values())
   avg_ratio_in = total_p_in / (total_k_in + EPSILON)
   total_p_out, total_k_out = sum(p_out.values()), sum(k_out.values())
   avg_ratio_out = total_p_out / (total_k_out + EPSILON)
   load_in, load_out = {}, {}
   for node in graph.nodes:
       # Safeguard against divide by zero and invalid logs
       p_in_value = max(p_in.get(node, 0), EPSILON)
       k_in_value = max(k_in[node], EPSILON)
       p_out_value = max(p_out.get(node, 0), EPSILON)
       k_out_value = max(k_out[node], EPSILON)
       load_in[node] = np.log((p_in_value / k_in_value) / (avg_ratio_in + EPSILON))
       load_out[node] = np.log((p_out_value / k_out_value) / (avg_ratio_out + EPSILON))
   return {"load_in": load_in, "load_out": load_out}
# 4. Centralities
def compute_centralities(graph):
   return {
       "degree": nx.degree_centrality(graph),
       "betweenness": nx.betweenness_centrality(graph),
       "closeness": nx.closeness_centrality(graph)
   }
# 5. Node-Based Choke Points
def find_node_based_choke_points(graph, load_data):
   choke_points = []
   for node in graph.nodes:
       if graph.degree[node] == 2:  # Check for degree 2
           neighbors = list(graph.neighbors(node))
           if len(neighbors) == 2:
               u, v = neighbors
               G_copy = graph.copy()
               G_copy.remove_node(node)
               if not nx.has_path(G_copy, u, v):
                   cost_impact = float('inf')  # Disconnection
               else:
                   original_length = nx.shortest_path_length(graph, u, v)
                   new_length = nx.shortest_path_length(G_copy, u, v)
                   cost_impact = (new_length + EPSILON) / (original_length + EPSILON)
               load_contribution = load_data["load_in"].get(node, 0) + load_data["load_out"].get(node, 0)
               choke_points.append((node, cost_impact, load_contribution))
   return sorted(choke_points, key=lambda x: (-x[1], -x[2]))
# 6. Visualization for Choke Points and Load Points
def visualize_graph(graph, choke_points, load_data, output_file):
   pos = nx.spring_layout(graph)
   total_load = {node: load_data["load_in"][node] + load_data["load_out"][node] for node in graph.nodes}
   threshold = np.percentile(list(total_load.values()), 80)
   load_points = [node for node, load in total_load.items() if load >= threshold]
   node_colors = []
   for node in graph.nodes:
       if node in [cp[0] for cp in choke_points]:
           node_colors.append("red")  # Choke points
       elif node in load_points:
           node_colors.append("orange")  # Load points
       else:
           node_colors.append("blue")  # Other nodes
   nx.draw(graph, pos, with_labels=True, node_color=node_colors, font_weight="bold", node_size=700)
   plt.title("Graph with Load Points and Choke Points", fontsize=16)
   plt.savefig(output_file, dpi=300)
   plt.show()
```

# Example Graph: Traffic Network

```python
# Example Graph: Traffic Network
G = nx.DiGraph()
city_map = {1: "City A", 2: "City B", 3: "City C", 4: "City D", 5: "City E"}
G.add_weighted_edges_from([
   (1, 2, 5), (2, 3, 10), (3, 4, 3), (4, 5, 8),
   (1, 3, 15), (2, 4, 20), (1, 5, 25)
])
# Relabel nodes before computing metrics
G = nx.relabel_nodes(G, city_map)
# Compute Metrics
p_in, p_out = compute_path_counts(G, k=3)
k_in, k_out = {n: G.in_degree(n) for n in G.nodes}, {n: G.out_degree(n) for n in G.nodes}
load_data = compute_load(G, p_in, k_in, p_out, k_out)
centralities = compute_centralities(G)
node_choke_points = find_node_based_choke_points(G.to_undirected(), load_data)
# Visualize Graph
visualize_graph(G, node_choke_points, load_data, "graph_with_load_and_choke_points.png")
# Output Results
load_matrix = pd.DataFrame(load_data)
centrality_matrix = pd.DataFrame(centralities)
choke_matrix = pd.DataFrame(node_choke_points, columns=["Node", "Cost Impact", "Load Contribution"])
# Print results
print("Load Matrix:")
print(load_matrix)
print("\nCentrality Matrix:")
print(centrality_matrix)
print("\nNode-Based Choke Points:")
print(choke_matrix)
```

# Example: Drug Target Identification in a Metabolic Network

In a metabolic network, we can represent metabolites as nodes, and enzymes/genes as connecting entities (edges). Choke points with light load (low traffic but critical for connectivity) can be prioritized as potential drug targets since inhibiting them will disrupt the pathway while minimizing unintended effects on the network.
Example Network
Structure:
- Nodes: Metabolites .
- Edges: Enzymes  or Genes  connecting metabolites.
- Weights: Reaction flux, connection strength, or number of pathways utilizing the edge.
---

## Graph Construction

```python

# Example Metabolic Network
metabolic_graph = nx.DiGraph()
# Add nodes (metabolites)
metabolic_graph.add_nodes_from(["M1", "M2", "M3", "M4", "M5"])
# Add edges (reactions catalyzed by enzymes or genes)
# Edge weights represent reaction flux or importance
metabolic_graph.add_weighted_edges_from([
   ("M1", "M2", 5),  # Enzyme E1
   ("M2", "M3", 10), # Enzyme E2
   ("M3", "M4", 3),  # Enzyme E3
   ("M4", "M5", 8),  # Enzyme E4
   ("M1", "M3", 15), # Enzyme E5
   ("M2", "M4", 20), # Enzyme E6
   ("M1", "M5", 25)  # Enzyme E7
])
```

Applying the Algorithms

```python

# 1. Compute Load and Centralities
# Compute path counts for k-shortest paths
p_in, p_out = compute_path_counts(metabolic_graph, k=3)
# Compute neighbor counts
k_in = {n: metabolic_graph.in_degree(n) for n in metabolic_graph.nodes}
k_out = {n: metabolic_graph.out_degree(n) for n in metabolic_graph.nodes}
# Compute load
load_data = compute_load(metabolic_graph, p_in, k_in, p_out, k_out)
# Compute centralities
centralities = compute_centralities(metabolic_graph)
# 2. Identify Choke Points
# Find node-based choke points
node_choke_points = find_node_based_choke_points(metabolic_graph.to_undirected(), load_data)
# 3. Highlight Light Load Choke Points
# Extract load values and identify light load choke points
total_load = {node: load_data["load_in"][node] + load_data["load_out"][node] for node in metabolic_graph.nodes}
light_load_choke_points = [
   (node, load, centralities["betweenness"].get(node, 0))
   for node, load in total_load.items()
   if load < np.percentile(list(total_load.values()), 50)  # Below median load
   and node in [cp[0] for cp in node_choke_points]
]
# Visualizing the Network
def visualize_metabolic_network(graph, choke_points, light_load_points, output_file):
   """
   Visualizes the metabolic network highlighting choke points and light-load points.
   Parameters:
       graph: A NetworkX graph.
       choke_points: List of choke points.
       light_load_points: List of nodes with light load.
       output_file: Filename to save the visualization.
   """
   pos = nx.spring_layout(graph)
   # Node coloring based on type
   node_colors = []
   for node in graph.nodes:
       if node in [cp[0] for cp in choke_points]:
           node_colors.append("red")  # Choke points
       elif node in [lp[0] for lp in light_load_points]:
           node_colors.append("green")  # Light load points
       else:
           node_colors.append("blue")  # Other nodes
   nx.draw(graph, pos, with_labels=True, node_color=node_colors, node_size=700, font_weight="bold")
   plt.title("Metabolic Network: Choke Points and Light-Load Targets", fontsize=16)
   plt.savefig(output_file, dpi=300)
   plt.show()
# Visualize the metabolic network
visualize_metabolic_network(
   metabolic_graph,
   node_choke_points,
   light_load_choke_points,
   "metabolic_network_targets.png"
)
# Output Results
# 1. Light-Load Choke Points
# Display light-load choke points
import pandas as pd
light_load_df = pd.DataFrame(
   light_load_choke_points,
   columns=["Node (Metabolite)", "Load Contribution", "Betweenness Centrality"]
)
print("Light-Load Choke Points (Potential Drug Targets):")
print(light_load_df)
```
